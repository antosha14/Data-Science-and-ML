## 1) Histogram with Frequency poligon (Классические столбики)

```python
data = pd.Series(np.random.standard_normal(100))
plt.hist(data, bins=20, color="black", alpha=0.3)
data.plot.density()
#sns.histplot(data,bins=20, color="black")
```

## 2) Сumulative frequency distribution (накопление по распределению)

```python
data = np.random.standard_normal(50).cumsum()
plt.plot(data, color="black", linestyle="dashed")
```

## 3) BAR CHART (горизонтальные столбики)

```python
df = pd.DataFrame(
    np.random.uniform(size=(6, 4)),
    index=["one", "two", "three", "four", "five", "six"],
    columns=pd.Index(["A", "B", "C", "D"], name="Genus"),
)
df.plot.barh()
sns.barplot(data=df, orient="h")
```

## 4) Grouped bar chart (несколько сгруппированных горизонтальных)

```python
df = pd.DataFrame(
    np.random.uniform(size=(6, 4)),
    index=["one", "two", "three", "four", "five", "six"],
    columns=pd.Index(["A", "B", "C", "D"], name="Genus"),
)
df.plot.barh(stacked=True, alpha=0.5)
```

## 5) Stacked bar chart (застаканные столбики)

```python
df = pd.DataFrame(
    np.random.uniform(size=(6, 4)),
    index=["one", "two", "three", "four", "five", "six"],
    columns=pd.Index(["A", "B", "C", "D"], name="Genus"),
)
df.plot.bar(stacked=True, alpha=0.5)
```
